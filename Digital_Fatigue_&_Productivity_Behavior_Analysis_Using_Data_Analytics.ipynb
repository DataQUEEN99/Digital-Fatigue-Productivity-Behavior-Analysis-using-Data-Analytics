```python

# DIGITAL FATIGUE ANALYTICS ENGINE


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

sns.set(style="whitegrid", palette="Set2")

# Upload CSV
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Basic Cleaning
df = df.drop_duplicates()
for col in df.select_dtypes(include=np.number):
    df[col].fillna(df[col].median(), inplace=True)

# Feature Engineering
df['screen_sleep_ratio'] = df['daily_screen_time_hours'] / df['sleep_hours']
df['work_efficiency'] = df['productivity_score'] / df['work_study_hours']


#  PRODUCTIVITY DISTRIBUTION
plt.figure(figsize=(10,5))
sns.histplot(df['productivity_score'], kde=True, color='purple')
plt.title("Overall Productivity Score Distribution", fontsize=14)
plt.show()


# VISUAL 2: SCREEN TIME vs PRODUCTIVITY
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='daily_screen_time_hours',
    y='productivity_score',
    hue='stress_level',
    palette='coolwarm',
    data=df
)
plt.title("Screen Time vs Productivity (Colored by Stress Level)")
plt.show()


#  STRESS IMPACT

plt.figure(figsize=(10,5))
sns.boxplot(
    x='stress_level',
    y='productivity_score',
    data=df,
    palette='Spectral'
)
plt.title("Impact of Stress Level on Productivity")
plt.show()

#  SLEEP vs PRODUCTIVITY

plt.figure(figsize=(10,5))
sns.lineplot(
    x='sleep_hours',
    y='productivity_score',
    data=df,
    ci=None,
    color='green'
)
plt.title("Sleep Duration vs Productivity Trend")
plt.show()


#  CORRELATION HEATMAP

plt.figure(figsize=(12,7))
sns.heatmap(
    df.corr(),
    annot=False,
    cmap='RdYlBu',
    linewidths=0.5
)
plt.title("Digital Behavior Correlation Matrix")
plt.show()
```
